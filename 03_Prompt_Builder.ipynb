# 03 Prompt Builder
Constructs the therapeutic prompt.

```python
def get_therapist_persona():
    return (
        "You are a highly empathetic, patient, and professional AI therapist. "
        "Your therapeutic approach is based on Person-Centered Therapy and Active Listening. "
        "Guidelines:\n"
        "1. Validation First: Acknowledge feelings.\n"
        "2. Reflective Listening: Use 'It sounds like...'.\n"
        "3. Personalization: Reference user profile context.\n"
        "4. Emotional Support: Adapt tone to mood.\n"
        "5. Gentle Inquiry: Ask open-ended questions."
    )

def build_final_prompt(user_text, recent_history, long_term_memory, analysis_data):
    prompt = f"## ROLE\n{get_therapist_persona()}\n\n"
    
    name = long_term_memory.get("user_name", "User")
    notes = long_term_memory.get("persistent_notes", [])
    
    prompt += f"## USER PROFILE\nName: {name}\n"
    if notes: prompt += "Past Insights:\n- " + "\n- ".join(notes[-5:]) + "\n"
    
    tone = analysis_data.get('emotional_tone', 'neutral')
    sentiment = analysis_data.get('sentiment', 'neutral')
    prompt += f"\n## STATE\nTone: {tone}, Sentiment: {sentiment}\n"
    
    prompt += "\n## HISTORY\n"
    for turn in recent_history:
        r = "Therapist" if turn['role'] == "assistant" else "User"
        prompt += f"{r}: {turn['content']}\n"
    
    prompt += f"\nUser: {user_text}\nTherapist: "
    return prompt
```
