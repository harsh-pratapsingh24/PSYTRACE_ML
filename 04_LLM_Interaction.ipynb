# 04 LLM Interaction Module (Groq Powered)
This module handles communication with the Groq API. It uses the latest Llama models for lightning-fast responses.

```python
import os
from groq import Groq
from dotenv import load_dotenv

def initialize_api():
    """
    Initializes the Groq client using the API key from .env.
    """
    cwd_env = os.path.join(os.getcwd(), '.env')
    if os.path.exists(cwd_env): load_dotenv(dotenv_path=cwd_env, override=True)
    else: load_dotenv()
        
    api_key = os.getenv("GROQ_API_KEY")
    if api_key:
        return Groq(api_key=api_key)
    return None

def get_mock_response(prompt):
    if "anxious" in prompt.lower(): return "It sounds like you're carrying a lot of worry. I'm here."
    return "I hear you. Tell me more."

def get_ai_response(prompt):
    client = initialize_api()
    
    if not client:
        return get_mock_response(prompt)
    
    try:
        # Using 'llama-3.3-70b-versatile' as the primary stable model
        # Fallback to 'llama3-70b-8192' if needed
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                }
            ],
            model="llama-3.3-70b-versatile",
        )
        return chat_completion.choices[0].message.content.strip()
    except Exception as e:
        # If the versatile model fails, try the standard 70b as a secondary fallback
        try:
             chat_completion = client.chat.completions.create(
                messages=[{"role": "user", "content": prompt}],
                model="llama3-70b-8192",
             )
             return chat_completion.choices[0].message.content.strip()
        except:
             return f"[Groq API Error] {str(e)}"
```
