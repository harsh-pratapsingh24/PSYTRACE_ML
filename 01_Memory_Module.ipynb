# 01 Memory Module
This module handles the storage and retrieval of conversation history and user information.

```python
# Global state for memory
conversation_history = []
user_profile = {
    "name": "Unknown",
    "key_topics": [],
    "mood_trend": []
}
```

```python
def save_interaction(role, text):
    """Saves a single turn to the conversation history."""
    conversation_history.append({"role": role, "content": text})

def get_recent_history(num_turns=5):
    """Retrieves the last N interactions for context."""
    return conversation_history[-num_turns:]

def update_user_profile(key, value):
    """Updates specific fields in the user profile."""
    if key in user_profile:
        if isinstance(user_profile[key], list):
            user_profile[key].append(value)
        else:
            user_profile[key] = value
```

```python
# Example usage (for testing within this notebook)
if __name__ == "__main__":
    save_interaction("user", "I feel a bit overwhelmed today.")
    save_interaction("assistant", "I'm sorry to hear that. What's on your mind?")
    print("History:", get_recent_history())
    update_user_profile("name", "Alex")
    print("Profile:", user_profile)
```
