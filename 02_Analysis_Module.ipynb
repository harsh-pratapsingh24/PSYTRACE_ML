# 02 Analysis Module
Performs NLP analysis on user input.

```python
from textblob import TextBlob

def analyze_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    if polarity > 0.1: return "positive"
    elif polarity < -0.1: return "negative"
    return "neutral"

def detect_emotional_tone(text):
    text_lower = text.lower()
    rules = {
        "stressed": ["stressed", "overwhelmed", "pressure", "busy"],
        "anxious": ["anxious", "worried", "fear", "nervous", "panic"],
        "sad": ["sad", "unhappy", "lonely", "depressed"],
        "happy": ["happy", "great", "excited", "wonderful"],
        "angry": ["angry", "mad", "annoyed", "frustrated"],
        "calm": ["calm", "relaxed", "peaceful"]
    }
    for tone, keywords in rules.items():
        if any(kw in text_lower for kw in keywords): return tone
    return "neutral"

def detect_risk(text):
    risk_keywords = ["hurt myself", "suicide", "kill myself", "help me"]
    text_lower = text.lower()
    return any(kw in text_lower for kw in risk_keywords)

def extract_psychological_insights(text):
    """Extracts emotions, patterns, and concerns to store in Long-term Memory."""
    insights = {"emotions": [], "patterns": [], "concerns": []}
    text_lower = text.lower()
    
    # Emotion keywords
    e_kws = {"sadness": ["sad", "lonely"], "anxiety": ["anxious", "worried"], "stress": ["stressed"]}
    # Pattern keywords
    p_kws = {"sleep_issue": ["can't sleep", "insomnia"], "social_withdrawal": ["avoiding people"]}
    # Concern keywords
    c_kws = {"work": ["job", "boss", "career"], "family": ["mom", "dad", "parents"]}
    
    for e, kws in e_kws.items():
        if any(kw in text_lower for kw in kws): insights["emotions"].append(e)
    for p, kws in p_kws.items():
        if any(kw in text_lower for kw in kws): insights["patterns"].append(p)
    for c, kws in c_kws.items():
        if any(kw in text_lower for kw in kws): insights["concerns"].append(c)
        
    return insights if any(insights.values()) else None

def run_full_analysis(text):
    return {
        "sentiment": analyze_sentiment(text),
        "emotional_tone": detect_emotional_tone(text),
        "is_high_risk": detect_risk(text),
        "insights": extract_psychological_insights(text)
    }
```
