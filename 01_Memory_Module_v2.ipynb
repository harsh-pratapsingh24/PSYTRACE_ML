# 01 Memory Module
Handles short-term and long-term memory persistence.

```python
import json
import os

MEMORY_FILE = 'memory.json'
MAX_SHORT_TERM_SIZE = 10

short_term_history = []
long_term_data = {
    "user_name": "User",
    "persistent_notes": [],
    "session_count": 0
}

def load_long_term_memory():
    global long_term_data
    if os.path.exists(MEMORY_FILE):
        try:
            with open(MEMORY_FILE, 'r') as f:
                long_term_data = json.load(f)
        except:
            pass

def save_long_term_memory():
    try:
        with open(MEMORY_FILE, 'w') as f:
            json.dump(long_term_data, f, indent=4)
    except:
        pass

def add_to_history(role, content):
    global short_term_history
    short_term_history.append({"role": role, "content": content})
    if len(short_term_history) > MAX_SHORT_TERM_SIZE:
        short_term_history = short_term_history[-MAX_SHORT_TERM_SIZE:]

def update_note(note_text):
    long_term_data["persistent_notes"].append(note_text)
    save_long_term_memory()
```
